```python
# Function to calculate % of missing values
def missing_values_table(df):
  mis_val = df.isnull().sum()
  
  mis_val_percent = 100 * df.isnull().sum() / df.shape[0]
  
  # make a table with the results
  mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

  # Rename the columns
  new_table = mis_val_table.rename(columns = {0:'Missing Values', 1: '% of Total Values'})
  
  # Sort the values by percentage of missing in descending order
  new_table = new_table[new_table.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False).round(1)
  
  # Summary of our dataframe
  print("Your selected dataframe has " + str(df.shape[1]) + " columns.\n" "There are "+ str(new_table.shape[0]) 
        + " columns that have missing values.")
  
  return new_table
```

```python
# Label Encoding and One Hot Encoding
# we use LE for 2 categorical values and OHE for more than 2 values
# create Label Encoder object
le = LabelEncoder()
le_count = 0

# Iterate through columns
for col in train:
  if train[col].dtype == 'object':
    # if 2 or fewer unique categories
    if len(list(train[col].unique())) <= 2:
    # train on the training data  
      le.fit(train[col])
      
    # transform both training and testing data
      train[col] = le.transform(train[col])
      test[col] = le.transform(test[col])
      
    # keep track of how many columns were label encoded
      le_count += 1
      
print("%d columns were label encodes." %le_count)
```

```python
# one hot encoding of categorical
train = pd.get_dummies(train)
test = pd.get_dummies(test)

print("Training shape: ", train.shape)
print("Testing shape: ", test.shape)
```

One hot encoding crated same features in both training and testing data. One-hot encoding has created more columns because there were some categorical variables with categories not represented in the testing data. To remove columns in the training data that are note in the testing data, we need to align the dataframes. First we extract the target column from the training data.

```python
train_labels = train["TARGET"]

train, test = train.align(test, join='inner',axis=1)

# add TARGET variable back into training set
train["TARGET"] = train_labels

print("Training shape: ", train.shape)
print("Testing shape: ", test.shape)
```

## Correlations
Let's find out the relation between the features and **Target**.
We will use **Pearson correlation** between every variable and the target.
It's not considered as greatest method to find 'relevance' of a feature, but it does give us an idea of possible relationship within the data. we can use below criteria to define how strong the correlation is.
* .00 - .19   -->  very weak
* .20 - .39   -->  weak
* .40 - .59   -->  moderate
* .60 - .79   -->  strong
* .80 - 1.0   -->  very strong

```python
correlations = train.corr()['TARGET'].sort_values()

print("Most positive correlation:\n", correlations.tail(10))
print("\nMost Negative correlation:\n", correlations.head(10))
```

With this results, we can say that `DAYS_BIRTH` and `DAYS_EMPLOYED` are most positive correlation. `DAYS_BIRTH` is the age in days of the client at the time of the loan in negative days. The correlation is positive, but the value of this feature is actually negative which means as the person(customer) gets older, they are less likely to default on their loan (i.e target == 0).

## Feature Engineering

We will create two simple feature engineering methods:
1. Polynomial Featues
2. Domain knowledge features

### 1. Polynomial Features
We will create a polynomial of existing features. The concept behind to create this features is that, two variables by themselves may not have a strong influence on the target, combining them together into a single variable might have a better relationship with the target. 

We will use Scikit-Learn's `PolynomialFeatures` that creates the polynomials and the interaction terms up to a specified degree. 

```python
# New DataFrame for polynomial features
pol_features = train[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3','DAYS_BIRTH','TARGET']]
pol_features_test = test[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3','DAYS_BIRTH']]

# Use imputer to handle missing values
from sklearn.preprocessing import Imputer
imputer = Imputer(strategy = 'median')

pol_target = pol_features['TARGET']
pol_features = pol_features.drop(columns = ['TARGET'])

# Impute the missing values
pol_features = imputer.fit_transform(pol_features)
pol_features_test = imputer.transform(pol_features_test)

# Create the polynomial features with specified degree
from sklearn.preprocessing import PolynomialFeatures

pol_transform = PolynomialFeatures(degree = 3)

pol_features = pol_transform.fit_transform(pol_features)
pol_features_test = pol_transform.fit_transform(pol_features_test)

print("Polynomial Features shape: ", pol_features.shape)
```

To get the names of new features, we need to use polynomial features `get_features_names` method.

```python
# we will get all 35 features
pol_transform.get_feature_names(input_features=['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3', 'DAYS_BIRTH'])
```

```python
# finding the correlation of this 35 features with the 'TARGET' variable
pol_features = pd.DataFrame(pol_features, columns=pol_transform.get_feature_names(input_features=['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3', 'DAYS_BIRTH']))

pol_features['TARGET'] = pol_target

# Find the correlation with the 'TARGET'
pol_corr = pol_features.corr()['TARGET'].sort_values()

# print most negative and most positive correlation
print(pol_corr.head(10))
print(pol_corr.tail(10))
```

```python
# Put this features on test DataFrame
pol_features_test = pd.DataFrame(pol_features_test, columns=pol_transform.get_feature_names(['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3', 'DAYS_BIRTH']))

# Merge polynomial features into training DataFrame
pol_features['SK_ID_CURR'] = train['SK_ID_CURR']
pol_train = train.merge(pol_features, on = 'SK_ID_CURR', how = 'left')

# Merge polynomial features into testing DataFrame
pol_features_test['SK_ID_CURR'] = test['SK_ID_CURR']
pol_test = test.merge(pol_features_test, on = 'SK_ID_CURR', how = 'left')
```

```python
print(pol_train.shape), print(pol_test.shape)
```

```python
# We need to align both the dataframe
pol_train, pol_test = pol_train.align(pol_test, join = 'inner', axis = 1)

# print the new shapes
print("Shape of Training data with polynomial features: ", pol_train.shape)
print("Shape of Testing data with polynomial features: ", pol_test.shape)
```

## Domain Knowledge Features
we will use five features as described below:
1. CREDIT_INCOME_PERCENT: The percentage of the credit amount relative to a client's income
2. ANNUITY_INCOME_PERCENT: The percentage of the loan annuity relative to a client's income
3. CREDIT_TERM: the length of the payment in months
4. DAYS_EMPLOYED_PERCENT: the percentage of the days employed relative to the client's age

```python
train_domain = train.copy()
test_domain = test.copy()

train_domain['CREDIT_INCOME_PERCENT'] = train_domain['AMT_CREDIT'] / train_domain['AMT_INCOME_TOTAL']
train_domain['ANNUITY_INCOME_PERCENT'] = train_domain['AMT_ANNUITY'] / train_domain['AMT_INCOME_TOTAL']
train_domain['CREDIT_TERM'] = train_domain['AMT_ANNUITY'] / train_domain['AMT_CREDIT']
train_domain['DAYS_EMPLOYED_PERCENT'] = train_domain['DAYS_EMPLOYED'] / train_domain['DAYS_BIRTH']
```

```python
test_domain['CREDIT_INCOME_PERCENT'] = test_domain['AMT_CREDIT'] / test_domain['AMT_INCOME_TOTAL']
test_domain['ANNUITY_INCOME_PERCENT'] = test_domain['AMT_ANNUITY'] / test_domain['AMT_INCOME_TOTAL']
test_domain['CREDIT_TERM'] = test_domain['AMT_ANNUITY'] / test_domain['AMT_CREDIT']
test_domain['DAYS_EMPLOYED_PERCENT'] = test_domain['DAYS_EMPLOYED'] / test_domain['DAYS_BIRTH']
```

```python
# Let's visualize New domain variables against TARGET variable
plt.figure(figsize=(12,20))

# iterate through the new features
for i, feature in enumerate(['CREDIT_INCOME_PERCENT','ANNUITY_INCOME_PERCENT','CREDIT_TERM', 'DAYS_EMPLOYED_PERCENT']):
    
    # create a new subplot for each source
    plt.subplot(4, 1, i + 1)
    
    # plot the repaid loans
    sns.kdeplot(train_domain.loc[train_domain['TARGET'] == 1, feature], label = "repaid")
    sns.kdeplot(train_domain.loc[train_domain['TARGET'] == 0, feature], label = 'not paid')
    
    # labelling
    plt.title("Distribution of %s by Target Value" % feature)
    plt.xlabel("%s" %feature)
    plt.ylabel("Density")
    
plt.tight_layout(h_pad = 2.5)
```

Now we will make some baseline model which would be our quick and sample base results. 

We will use simple classification algorithm called **Logistic Regression**. first we will pre-process the data by filling in the missing values (Imputation) and Normalizing the range of features (feature scaling).

```python
from sklearn.preprocessing import MinMaxScaler, Imputer

# Drop the 'TARGET' and make a copy of that dataframe
new_train = train.drop(columns=['TARGET'])
new_test = test.copy()

# Will use "median" imputer strategy
imputer = Imputer()

# Scale each feature with "mean = 0" and "standard deviation = 1"
scaler = MinMaxScaler(feature_range = (0, 1))

# Fit imputer on training data
imputer.fit(new_train)

# Transforming both "Train" and "Test" dataset
new_train = imputer.transform(new_train)
new_test = imputer.transform(new_test)

# Applying scaling on both
scaler.fit(new_train)
new_train = scaler.transform(new_train)
new_test = scaler.transform(new_test)
```

```python
print("Training shape: ", new_train.shape)
print("Testing shape: ", new_test.shape)
```

We need to predict probabilities (not only `0` and `1`), so we will use `predict_proba` to estimate its probabilities.

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(C = 0.0001)

# Train the data
log_reg.fit(new_train, train_labels)
```

Here, we want to predict the probabilities of not paying a loan, so we use the model `predict_proba` method. This will return an n x 2 array where n is the number of observations. First column is the probability of the `TARGET` = 0 and the second column is the probability of the `TARGET`  = 1. We want to find the probability of the loan is not paid, we we will only need second column.

```python
# Making prediction using second column
log_reg_pred = log_reg.predict_proba(new_test)[:, 1]
```

```python
preds = test[['SK_ID_CURR']]
preds['TARGET'] = log_reg_pred

preds.head()
```
